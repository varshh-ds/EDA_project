```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = df.dropna()
#The drop() method takes a list of column names in square brackets [] inside parentheses (), and axis=1 specifies weâ€™re dropping columns.
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(['id', 'key'], axis = 1)
```

```python
df['duration_sec'] = df['duration_ms'] / 1000  # Convert duration to seconds
```

```python
plt.figure(figsize = (8, 6))
sb.histplot(df['danceability'])
plt.title('Distribution of Danceability')
plt.xlabel('Danceability (0 to 1)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figure = (7, 6))
sb.histplot(df['streams'])
plt.title('Streaming')
plt.xlabel('streams')
plt.show()
```

```python
df['mode'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Count of Songs by Mode')
plt.xlabel('Mode (0 = Minor, 1 = Major)')
plt.ylabel('Count')
plt.show()
```

```python
print("Danceability Stats:\n", df['danceability'].describe())
print("\nStreams Stats:\n", df['streams'].describe())
print("\nMode Counts:\n", df['mode'].value_counts())
```

```python
viral_threshold = df['streams'].quantile(0.90)
df['is_viral'] = df['streams'] > viral_threshold
```

```python
features = ['danceability', 'energy', 'valence', 'tempo']

for feature in features:
    sb.boxplot(x='is_viral', y=feature, data=df)
    plt.title(f'{feature.capitalize()} by Viral Status')
    plt.xlabel('Is Viral')
    plt.ylabel(feature.capitalize())
    plt.show()
```

```python
""" Danceability by Viral Status
Observation: Viral songs and non-viral songs have similar danceability ranges and medians.

Interpretation: Danceability alone doesn't drastically differentiate viral songs â€” both viral and non-viral ones can be equally danceable.

âœ… Energy by Viral Status
Observation: Viral and non-viral songs have very similar energy distributions.

Interpretation: Energy (i.e., intensity, activity) is not a strong predictor of virality by itself.

âœ… Valence by Viral Status
Observation: Both groups show a wide range of valence (musical positivity). Medians are also similar.

Interpretation: Viral songs are not necessarily happier or more positive. Again, valence alone doesnâ€™t determine virality.

âœ… Tempo by Viral Status
Observation: Viral and non-viral songs cover a similar range of tempos. Medians are close.

Interpretation: Faster tempo doesn't guarantee virality. Songs with a wide variety of tempos can go viral."""
```

```python
dance_thresh = df['danceability'].median()
energy_thresh = df['energy'].median()

both_high_viral = df[(df['is_viral'] == 1) & 
                     (df['danceability'] > dance_thresh) & 
                     (df['energy'] > energy_thresh)].shape[0]

both_high_nonviral = df[(df['is_viral'] == 0) & 
                        (df['danceability'] > dance_thresh) & 
                        (df['energy'] > energy_thresh)].shape[0]

print(f'High Danceability & High Energy: Viral = {both_high_viral}, Non-Viral = {both_high_nonviral}')
```

```python
sb.scatterplot(data=df, x='danceability', y='energy', hue='is_viral', alpha=0.6)
plt.title('Danceability vs Energy by Viral Status')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
sb.scatterplot(data=df, x='valence', y='tempo', hue='is_viral', alpha=0.6)
plt.title('Valence vs Tempo by Viral Status')
plt.xlabel('Valence')
plt.ylabel('Tempo')
plt.show()
```

```python
#####ok
dance_thresh = df['danceability'].quantile(0.75)  # top 25%
energy_thresh = df['energy'].quantile(0.75)      # top 25%

viral_extremes = df[(df['is_viral'] == 1) & 
                    (df['danceability'] > dance_thresh) & 
                    (df['energy'] > energy_thresh)]

print("Viral songs with high danceability & energy:")
print(viral_extremes[['danceability', 'energy', 'valence', 'tempo']].head())
print(f"Count: {viral_extremes.shape[0]}")
```

```python
valence_thresh = df['valence'].quantile(0.75)
tempo_thresh = df['tempo'].quantile(0.75)

viral_extremes_val_tempo = df[(df['is_viral'] == 1) &
                              (df['valence'] > valence_thresh) &
                              (df['tempo'] > tempo_thresh)]

print("Viral songs with high valence & tempo:")
print(viral_extremes_val_tempo[['danceability', 'energy', 'valence', 'tempo']].head())
print(f"Count: {viral_extremes_val_tempo.shape[0]}")
```

```python
print("Average features of viral extremes (danceability & energy):")
print(viral_extremes.select_dtypes(include='number').mean())

print("\nAverage features of viral extremes (valence & tempo):")
print(viral_extremes_val_tempo.select_dtypes(include='number').mean())
```
